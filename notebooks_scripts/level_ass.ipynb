```python
import sys
import os

# Add the project root to Python path
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))


from app.workflow.utils import read_files_recursive

from pandas.errors import SettingWithCopyWarning

from app.transformers.data_modifiers.mapper import StyleMapper
from app.transformers.data_modifiers.namespace import SkiStyle, SkierLevel, Slope
```

```python
def save_selected_columns_to_csv(df_selected_columns, output_path):
    output_path = output_path.replace("../data/", "../data/with_styles/")
    directory = os.path.dirname(output_path)
    if not os.path.exists(directory):
        os.makedirs(directory)

    df_selected_columns.to_csv(output_path)
```

```python
labeled_data, file_paths = read_files_recursive(path = "../data", return_filenames=True)
```

## Iterate over dataset

```python
#_current_index = 57
#del _current_index
```

```python
# Much simpler data iterator
if '_current_index' not in globals():
    _current_index = 0
else:
    _current_index = (_current_index + 1) % len(labeled_data)

# Get current dataframe and filename
current_data = labeled_data[_current_index]
current_file = file_paths[_current_index]

print(f"Showing file {_current_index}/{len(labeled_data)-1}: {current_file}")
print(f"Itr : {_current_index}")

print(current_data.index[0])
print(current_data.index[-1])
```

## Map data styles

```python
# Create a StyleMapper instance and apply it to the current dataframe
style_mapper = StyleMapper(
    style=SkiStyle.SKIDDED_SHORT,           # Choose from: CARVING, RACING, RECREATIONAL
    skier_level=SkierLevel.EXPERT,    # Choose from: BEGINNER, INTERMEDIATE, EXPERT
    slope=Slope.BLUE              # Choose from: FLAT, MODERATE, STEEP
)

# Apply mapping to current dataframe
mapped_data = style_mapper.map(current_data)

# Show the first few rows with the new columns
print(f"Data shape after mapping: {mapped_data.shape}")

# Keep the mapped data for further processing
current_data = mapped_data
current_data.head(5)
```

## SAVE

```python

save_selected_columns_to_csv(current_data, current_file)
print(f"Saved mapped data to {current_file} , iteration {_current_index}")
```
